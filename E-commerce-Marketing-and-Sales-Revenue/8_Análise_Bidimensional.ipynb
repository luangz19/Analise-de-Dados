## 2.3 - Padronizando os nomes da coluna region

```python
def cap(x):
  x = x.capitalize()
  if x == "Norht" or x == "Nort" or x == "North ":
    x = "North"
  return x

data_train["region"] = data_train["region"].apply(cap)
data_train["region"].value_counts()
```

```python
print(data_train["channel"].value_counts())
```

## 2.4 Padronizando os nomes da coluna channel

```python


def cap2(x):
  palavra = ""
  x = x.capitalize()
  for i in x:
    palavra += i
    if palavra == "Soc":
      x = "Social Media"
      break
      return x
  return x
```

```python
data_train["channel"] = data_train["channel"].apply(cap2)
data_train["channel"].value_counts()
```

# 10 - Análise Bidimensional

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(2,2, figsize=(15,8))
sns.boxplot(data_train, x="region", y="ad_spend", ax=ax[0,0])
sns.boxplot(data_train, x="region", y="price", ax=ax[0,1])
sns.boxplot(data_train, x="region", y="discount_rate", ax=ax[1,0])
sns.boxplot(data_train, x="region", y="market_reach", ax=ax[1,1])
plt.tight_layout()
```

```python
fig, ax = plt.subplots(2,2, figsize=(15,8))
sns.boxplot(data_train, x="region", y="click_through_rate", ax=ax[0,0])
sns.boxplot(data_train, x="region", y="competition_index", ax=ax[0,1])
sns.boxplot(data_train, x="region", y="seasonality_index", ax=ax[1,0])
sns.boxplot(data_train, x="region", y="campaign_duration_days", ax=ax[1,1])
plt.tight_layout()
```

```python
fig, ax = plt.subplots(2,1, figsize=(15,8))
sns.boxplot(data_train, x="region", y="customer_lifetime_value", ax=ax[0])
sns.boxplot(data_train, x="region", y="sales_revenue", ax=ax[1])
plt.tight_layout()
```

## 10.1 Matriz de Correlação


```python
fig, ax = plt.subplots(figsize=(15,8))
sns.heatmap(data_train[dados_quantitativos[1:]].corr("spearman"), annot=True, vmax=1, vmin=-1, ax=ax)
```

## 10.2 Matriz de Covariância

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.heatmap(data_train[dados_quantitativos[1:]].cov(), annot=True,  fmt=".2f", cmap="coolwarm", ax=ax)
```
