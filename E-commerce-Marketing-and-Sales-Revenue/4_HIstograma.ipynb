## 2.3 - Padronizando os nomes da coluna region

```python
def cap(x):
  x = x.capitalize()
  if x == "Norht" or x == "Nort" or x == "North ":
    x = "North"
  return x

data_train["region"] = data_train["region"].apply(cap)
data_train["region"].value_counts()
```

```python
print(data_train["channel"].value_counts())
```

## 2.4 Padronizando os nomes da coluna channel

```python


def cap2(x):
  palavra = ""
  x = x.capitalize()
  for i in x:
    palavra += i
    if palavra == "Soc":
      x = "Social Media"
      break
      return x
  return x
```

```python
data_train["channel"] = data_train["channel"].apply(cap2)
data_train["channel"].value_counts()
```

# 6 - Histogramas



## 6.1 - Histograma para Variáveis Qualitativas

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(2,2, figsize=(15,8))
sns.histplot(data_train["region"], ax=ax[0,0])
ax[0,0].set_title("Region")

sns.histplot(data_train["channel"], ax=ax[0,1])
ax[0,1].set_title("Channel")

sns.histplot(data_train["product_category"], ax=ax[1,0])
ax[1,0].set_title("Product Category")

sns.histplot(data_train["customer_segment"], ax=ax[1,1])
ax[1,1].set_title("Customer Regment")
plt.tight_layout()
```

## 5.2 - Histograma para Variáveis Quantitativas

```python
fig, ax = plt.subplots(2,2, figsize=(15,8))

sns.histplot(data_train["ad_spend"], ax=ax[0,0], bins=35)
ax[0,0].set_title("Gasto com Anúncio")

sns.histplot(data_train["price"], ax=ax[0,1], bins=35)
ax[0,1].set_title("Preço")

sns.histplot(data_train["discount_rate"], ax=ax[1,0], bins=20)
ax[1,0].set_title("Taxa de Desconto")

sns.histplot(data_train["market_reach"], ax=ax[1,1], bins=20)
ax[1,1].set_title("Alcance de Mercado")

plt.tight_layout()
```

```python
fig, ax = plt.subplots(2,2, figsize=(15,8))

sns.histplot(data_train["click_through_rate"], ax=ax[0,0], bins=35)
ax[0,0].set_title("Taxa de click")

sns.histplot(data_train["competition_index"], ax=ax[0,1], bins=35)
ax[0,1].set_title("Índice de Competição")

sns.histplot(data_train["seasonality_index"], ax=ax[1,0], bins=20)
ax[1,0].set_title("Índice de sazonalidade")

sns.histplot(data_train["campaign_duration_days"], ax=ax[1,1], bins=25)
ax[1,1].set_title("Duração da campanha em dias")

plt.tight_layout()
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,8))

sns.histplot(data_train["customer_lifetime_value"], ax=ax[0], bins=35)
ax[0].set_title("Valor do Ciclo de Vida do Cliente")

sns.histplot(data_train["sales_revenue"], ax=ax[1], bins=40)
ax[1].set_title("Receita de Vendas")

plt.tight_layout()
```
